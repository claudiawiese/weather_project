**Comparing different mlm models**

```python
# importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# loading dataset
df = pd.read_csv('mlm compare.csv')
display(df.head())
```

```python
# List of columns to plot
columns_to_plot = ['accuracy', 'precision', 'recall', 'F1']  # Add more columns as needed

# Create subplots
fig, axes = plt.subplots(nrows=len(columns_to_plot), ncols=1, figsize=(6, 4 * len(columns_to_plot)))

# Plot bar plots for each column
for i, column in enumerate(columns_to_plot):
    sns.barplot(x='model', y=column, data=df, ax=axes[i])
    axes[i].set_title(f'Barplot of {column}')
    axes[i].set_ylabel(column)

plt.tight_layout()
plt.savefig('mlm.png')
plt.show()
```
